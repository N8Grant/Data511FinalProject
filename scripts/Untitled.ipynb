```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_state = pd.read_excel('../data/CensusData/cleaned_state_data.xlsx')
df_state['State'] = df_state['State'].fillna("United States")
df_state.head(15)
```

```python
df_state.shape
```

list format
[State, Coverage, Year, Value]

Estimate,	Margin of error,	Percent,	Margin of error

```python
l = []
for state in df_state['State'].unique():
    state_data = df_state[df_state['State'] == state]
    year = 2019
    state_list = []
    for i in range(state_data.shape[0]):
        state_list.append(state)
        coverage_type = state_data.iloc[i,1]
        for j in range(2,state_data.shape[1]):
            if j % 2 == 0 and j % 4 != 0:
                l.append([state, coverage_type, 2020 - int(j/4),state_data.iloc[i,j]])
```

```python
l
```

```python
df3 = pd.DataFrame(data=l,columns=["State","Coverage","Year","Value"])
df3.to_csv("../data/number_insured.csv")
```

```python
race_df = pd.read_excel("../data/CensusData/coverage_2008_2019_race.xlsx")
race_df = race_df.iloc[2:,:]
race_df.head()
```

Array Setup
[race,year,percentage]

```python
l = []
counter = 0
for i in range(2,race_df.shape[0]-5):
    for j in range(4,61,5):
        l.append([race_df.iloc[i,0],2019-counter,100-float(race_df.iloc[i,j])])
        counter+=1
    counter = 0
```

```python
race_df_cleaned = pd.DataFrame(data=l, columns=["Race","Year","Percentage"])
race_df_cleaned.to_csv("../data/CensusData/race_data_formatted.csv")
```
